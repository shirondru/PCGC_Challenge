```python
import pandas as pd
```

```python
bad_var_df = pd.read_csv("no_ref_match_variants.csv")
```

```python
bad_var_df.head()
```

```python
#get var_id so I can remove these variants from analysis
bad_var_df['var_id'] = bad_var_df['chrom'].astype(str) + ':' + bad_var_df['pos'].astype(str) + ' ' + bad_var_df['ref'].astype(str) + '>' + bad_var_df['alt'].astype(str)
with open("no_ref_match_variants.txt",'w') as f:
    f.write("##These are variants whose reference and alt allele conflicted with what is reported on gnomAD, and were identified because their reference allele is different than the reference genome")
    f.write('\n')
    f.write(';'.join(bad_var_df['var_id'].tolist()))
```
